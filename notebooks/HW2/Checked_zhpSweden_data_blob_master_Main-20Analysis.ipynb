Next I create a class called HierarchicalPartion, which can hierarchically partition the stations into groups that have similar sum of weight. It also has the function called merge, which can merge two groups into one and update all the relevant fields. 

```python
import pandas as pd
import numpy as np

class HierarchicalPartion(object):

	def __init__(self, df, level):
		self.has_run = False
		self.df = df
		self.level = level
		self.partitioned_df = None
		self.leaves = None
		self.neighbor_matrix = None
		# a component may consist of one or more regions (leaf)
		self.components = {}
		self.id_generator = None

	def _generate_new_id(self):
		num = self.id_generator
		self.id_generator += 1
		return num

	def _get_id(self, bits):
		return sum(bit * (2**i) for bit, i in zip(bits, range(len(bits))))

	def _get_partition(self, partition):
		if partition == 'latitude':
			return 'longitude'
		return 'latitude'

	def _init_neighbors(self):
		num_leaves = len(self.leaves)
		self.neighbor_matrix = np.zeros((num_leaves, num_leaves), dtype=np.bool)
		for i in range(0, num_leaves):
			for j in range(i, num_leaves):
				if self._is_neighbor(i, j):
					self.neighbor_matrix[i][j] = True
					self.neighbor_matrix[j][i] = True

	def _equal(self, a, b):
		return np.absolute(a-b) < 1e-6

	def _double_check(self, x1, y1, x2, y2):
		if y1 < x2 or y2 < x1:
			return False
		return True

	def _is_neighbor(self, i, j):
		leaf_i = self.leaves.loc[i]
		leaf_j = self.leaves.loc[j]

		if self._equal(leaf_i.N, leaf_j.S) or self._equal(leaf_j.N, leaf_i.S):
			if self._double_check(leaf_i.W, leaf_i.E, leaf_j.W, leaf_j.E):
				return True

		if self._equal(leaf_i.E, leaf_j.W) or self._equal(leaf_j.E, leaf_i.W):
			if self._double_check(leaf_i.S, leaf_i.N, leaf_j.S, leaf_j.N):
				return True

		return False

	def _neighbors_of_region(self, i):
		num_leaves = len(self.leaves)
		neighbors = []
		for j in range(0, num_leaves):
			if self.neighbor_matrix[i][j]:
				neighbors.append(j)
		return neighbors

	def _neighbors_of_component(self, j):
		return self.components[j]['neighbors']

	def _are_components_neighbors(self, i, j):
		return i in self._neighbors_of_component(j)

	def merge(self, i, j):
		components = self.components
		if i in components and j in components and self._are_components_neighbors(i, j):
			c1 = components[i]
			c2 = components[j]
			weight_i = c1['weight']
			weight_j = c2['weight']
			region_contained_i = c1['region_contained']
			region_contained_j = c2['region_contained']
			neighbors_i = c1['neighbors']
			neighbors_j = c2['neighbors']

			new_id = self._generate_new_id()
			new_region_contained = list(set(region_contained_i) | set(region_contained_j))
			new_neighbors = list((set(neighbors_i) | set(neighbors_j)) - set([i,j]))
			new_weight = weight_i + weight_j

			components[new_id] = {'weight': new_weight,
			                      'region_contained': new_region_contained,
			                      'neighbors': new_neighbors}

			for n in new_neighbors:
				component = components[n]
				component['neighbors'] = [new_id] \
				                         + [n for n in component['neighbors']
				                            if n != i and n != j]
			del components[i]
			del components[j]

			df = self.partitioned_df
			df.loc[df.id == i, 'id'] = new_id
			df.loc[df.id == j, 'id'] = new_id

			print "successfully merge " + str(i) + "and " + str(j)
			return new_id
		else:
			print "merge failed"
			return -1

	def execute(self):
		if not self.has_run:
			self.has_run = True
			df = self.df.copy(deep=True)
			self.partitioned_df, leaves = self._execute(df, self.level)

			self.leaves = pd.DataFrame(leaves).set_index('id')
			self.id_generator = len(self.leaves)

			self._init_neighbors()
			for id, row in self.leaves.iterrows():
				self.components[id] = {'weight': row.weight,
				                       'region_contained': [id],
				                       'neighbors': self._neighbors_of_region(id)}

	def _execute(self, df, level, partition='latitude', N=90, S=-90, W=-180, E=180, bits=[]):
		if level <= 0:
			id = self._get_id(bits)
			df['id'] = id
			region = {'id': id,
			          'N': N, 'S': S, 'W': W, 'E': E,
			          'weight': df.weight.sum()}
			return df, [region]

		df = df.sort(partition)
		df['cumulative_weight'] = df.weight.cumsum()
		threshold = df.weight.sum()/2
		tree_id = 'tree' + str(level)
		df[tree_id] = df.cumulative_weight > threshold
		del df['cumulative_weight']
		grouped_df = df.groupby(tree_id)

		group_below_threshold = pd.DataFrame()
		group_above_threshold = pd.DataFrame()
		for flag, group in grouped_df:
			if flag: group_above_threshold = group
			else: group_below_threshold = group

		min_Value = group_above_threshold[partition].min()
		max_Value = group_below_threshold[partition].max()

		border = (max_Value + min_Value)/2

		children_df = []
		children_leaves = []

		for flag, child in grouped_df:
			if partition == 'latitude':
				if flag:
					S = border
				else:
					N = border
			else:
				if flag:
					W = border
				else:
					E = border
			child_df, child_leaves = self._execute(child,
			                                      level-1,
			                                      self._get_partition(partition),
			                                      N, S, W, E,
			                                      [flag]+bits)
			children_df.append(child_df)
			children_leaves += child_leaves

		return pd.concat(children_df), children_leaves
```

The output format of PCA is as follows:

[group_id, k, n], [[sum of all records] (a), the sum of all outer products of each records (b)]

We do it in this way because we need a and b to help us easily construct the covariance matrix of combined records of two groups. 
We can use the following formula to efficiently calculate the convariance matrix:

u3 = (a1+a2)/(n1+n2)
covariance matrix = (b1+b2 - np.outer((a1+a2), u3) - np.outer(u3, (a1+a2)) + np.outer(u3, u3))/(n1+n2)

We can use the output to see if it satisfies the merging condition.

```python
%%writefile pca.py
from mrjob.job import MRJob
import pickle
import pandas as pd
import numpy as np

class MRWeather(MRJob):

	def configure_options(self):
		super(MRWeather, self).configure_options()
		self.add_file_option('--metadata')

	def mapper_init(self):
		with open(self.options.metadata) as f:
			self.df = pickle.load(f)[['id']]

	def mapper(self, _, line):
		row = line.strip().split('\t')
		station = row[0][1:-1].split(', ')[0][1:-1]
		records = row[1][1:-1].split(', ')
		records = [int(record[1:-1]) for record in records]
		id = int(self.df.loc[station])
		yield id, records

	def reducer(self, id, values):
		data = pd.DataFrame(values, columns=range(1,5))
		data[data<-400] = np.nan
		data[data>500] = np.nan
		data = data/10
		Mean = np.mean(data, axis=0).values

		mat = data.transpose()
		cols, rows = mat.shape

		cov_mat = np.zeros((cols, cols))

		vector_sum = np.zeros(cols)
		cov = np.zeros((cols, cols))
		for i in range(rows):
			vector_sum += mat.iloc[:,i]
			cov += np.outer(mat.iloc[:,i], mat.iloc[:,i])
			row = mat.iloc[:,i] - Mean
			outer = np.outer(row, row)
			cov_mat += outer

		cov_mat /= rows

		U,D,V = np.linalg.svd(cov_mat)
		explained_var = np.cumsum(D)/np.sum(D)
		def binarySearch(arr, target):
			lo, hi = 0, len(arr)
			while lo <= hi:
				mid = lo + (hi-lo)/2
				if arr[mid] == target: return mid
				if arr[mid] < target: lo = mid + 1
				else: hi = mid - 1
			return lo

		k = binarySearch(explained_var, 0.99)+1
		n = rows

		yield (id, k, n), (list(vector_sum), [list(item) for item in cov])

if __name__ == '__main__':
	MRWeather.run()
```

### Merge Part ###

My merge strategy is heuristic and greedy. We start from the smallest id and get all its neighbors. We compute the k3 and n3 of each combined pair and pull out the one with the smallest n3*k3+(k3+1)*(2*365) and merge them. Then we put it into the list that represents the regions that have been already merged. When we meet any id that lies in the list, we just skip it and move to the next id. Then we do it this fashion until we finish the work for the last id. After each iteration, we get new groups, if we have merge any pairs in this iteration, it means we may potentially merge more, so we continue to merge until we find that we cannot merge any more. 

```python
import pickle
import numpy as np
from collections import defaultdict

def dump(obj, filename):
	with open(filename, 'w') as f:
		pickle.dump(obj, f)

def mr_pca(inputfile, df_file, outputfile):
	job_flow_id = ''
	#!python PCA.py $inputfile -r emr --emr-job-flow-id $job_flow_id --metadata $df_file > $outputfile

inputfile = ''
outputfile = ''

def parse(filename):
	hashmap = {}
	reg = u'\[.+?\]'
	import re
	reg = re.compile(reg)
	for line in open(filename):
		row = line.strip().split('\t')
		item1 = row[0][1:-1].split(', ')
		id = int(item1[0])
		k = int(item1[1])
		n = int(item1[2])
		data = row[1]
		index = data.find(']')
		vector_sum = data[1:index]
		vector_sum = [int(item) for item in vector_sum.split(', ')]
		cov = data[index+4:-1]
		cov_mat = []
		for item in reg.findall(cov):
			elems = item[1:-1].split(', ')
			cov_mat.append([int(i) for i in elems])
		hashmap[id] = {'k': k, 'n': n, 'vector_sum': np.array(vector_sum), 'cov_mat': np.array(cov_mat)}

	return hashmap

def formula(n, k):
	return n*k+(k+1)*(2*365)

def pca(v1, v2, c1, c2, n1, n2, k1, k2):
	new_sum = v1+v2
	Mean = new_sum/(n1+n2)
	cov_mat = c1+c2-np.outer(new_sum,Mean)-np.outer(Mean,new_sum)+np.outer(Mean,Mean)
	cov_mat /= n1+n2
	U,D,V = np.linalg.svd(cov_mat)
	explained_var = np.cumsum(D)/np.sum(D)
	def binarySearch(arr, target):
		lo, hi = 0, len(arr)
		while lo <= hi:
			mid = lo + (hi-lo)/2
			if arr[mid] == target: return mid
			if arr[mid] < target: lo = mid + 1
			else: hi = mid - 1
		return lo
	k3 = binarySearch(explained_var, 0.99)+1
	n3 = n1+n2
	left = formula(n1, k1) + formula(n2, k2)
	right = formula(n3, k3)
	if left > right:
		return True, right
	return False, -1

def get_attribute(hashmap, id):
	dictionary = hashmap[id]
	return dictionary['k'], dictionary['n'], dictionary['vector_sum'], dictionary['cov_mat']

def merge(hp, filename):
	hashmap = parse(filename)
	components = hp.components
	keys = sorted(components.keys())
	candidates = defaultdict(list)
	for i in range(len(keys)):
		for j in range(i+1, len(keys)):
			if hp._are_components_neighbors(keys[i], keys[j]):
				candidates[i].append(j)

	keys = sorted(candidates.keys())

	new_ones = []
	already_merged = []
	for id in keys:
		if id not in already_merged:
			res_pool = []
			id1 = id
			k1, n1, v1, c1 = get_attribute(hashmap, id1)
			for neighbor in candidates[id]:
				if neighbor not in already_merged:
					id2 = neighbor
					k2, n2, v2, c2 = get_attribute(hashmap, id2)
					flag, res = pca(v1, v2, c1, c2, n1, n2, k1, k2)
					if flag: res_pool.append([res, id2])
			if res_pool:
				res = min(res_pool)
				hp.merge(id1, res[1])
				new_ones.append(hp.id_generator-1)
				already_merged.append(res[1])

	return new_ones

def merge_loop(hp):
	epoch = 1
	while True:
		print "running " + str(epoch) + " epoch"
		df = hp.partitioned_df
		dump(df, 'metadata.pkl')
		mr_pca(inputfile, df, outputfile)
		new_ones = merge(hp, outputfile)
		epoch += 1
		if not new_ones:
			print "Terminate loop"
			break
```
